# Part 1 — Tuples

*Tuples are like lists, but **immutable** — once created, they cannot be changed. They are used when data should stay fixed, such as coordinates, database records, or lookup tables.*

## Question 1.1 — Written

What is the key difference between a **tuple** and a **list** in Python?  
Give one real-world data scenario where you would choose a tuple over a list, and explain why.

A tuple is an ordered collection of elements that is immutable, meaning its contents cannot be changed after creation. A list is also ordered, but it is mutable, so elements can be modified, added, or removed.

Real-world example:
A transaction record (order ID, date, amount) should be a tuple because it represents fixed data that should not be accidentally changed after being recorded.

## Question 1.2 — Code

Write code that:
1. Unpacks the tuple into individual named variables
2. Calculates and prints the total amount after discount
3. Prints a neat formatted summary of the record
4. Attempts to change the `city` value — observe what happens and write a comment explaining the error

````python
# This code unpacks a transaction tuple, calculates total cost, discount amount,
# and final payable amount, prints a formatted summary, and safely demonstrates tuple immutability

record = ("DT-20240001", "2024-03-15", "Lagos", "Electronics", "Wireless Headphones", 3, 49.99, 10)

# Unpacking the tuple
order_id, date, city, category, product, quantity, unit_price, discount_pct = record

# Calculations
total_cost = quantity * unit_price
discount_amount = total_cost * (discount_pct / 100)
final_total = total_cost - discount_amount

# Formatted output
print("----- ORDER SUMMARY -----")
print(f"Order ID: {order_id}")
print(f"Date: {date}")
print(f"City: {city}")
print(f"Category: {category}")
print(f"Product: {product}")
print(f"Quantity: {quantity}")
print(f"Unit Price: {unit_price}")
print(f"Total Cost: {total_cost:.2f}")
print(f"Discount (%): {discount_pct}")
print(f"Discount Amount: {discount_amount:.2f}")
print(f"Final Total: {final_total:.2f}")

# Demonstrating tuple immutability safely
try:
    record[2] = "Abuja"
except TypeError as e:
    print("\nTuple Immutability Demonstration:")
    print("Error:", e)
````

## Question 1.4 — Code ⭐ Challenge

Using the same `sales_records` list, write code that:
1. Groups the records by city using a dictionary (city → list of totals)
2. Calculates the total revenue per city
3. Prints a city revenue league table sorted from highest to lowest
4. Prints which city has the highest average order value

````python
# This code groups sales records by city, calculates total revenue per city,
# prints a sorted revenue league table, and identifies the city with the highest average order value

# Using sales_records from Question 1.3

city_data = {}

# 1. Group records by city (city → list of totals)
for order_id, city, category, amount in sales_records:
    if city not in city_data:
        city_data[city] = []
    city_data[city].append(amount)

# 2. Calculate total revenue per city
city_totals = {}
for city, amounts in city_data.items():
    city_totals[city] = sum(amounts)

# 3. Print league table (highest → lowest)
sorted_cities = sorted(city_totals.items(), key=lambda x: x[1], reverse=True)

print("----- City Revenue League Table -----")
for city, total in sorted_cities:
    print(f"{city}: ₦{total:.2f}")

# 4. Find city with highest average order value
highest_avg_city = ""
highest_avg = 0

for city, amounts in city_data.items():
    avg = sum(amounts) / len(amounts)
    if avg > highest_avg:
        highest_avg = avg
        highest_avg_city = city

print("\nCity with Highest Average Order Value:")
print(f"{highest_avg_city} (₦{highest_avg:.2f})")
````

# Part 2 — Dictionaries

*Dictionaries store data as key-value pairs. They are the most important Python structure for data work — a pandas DataFrame row is essentially a dictionary, and most data from APIs arrives as dictionaries (JSON).*

## Question 2.1 — Written

What is a dictionary in Python? How does it differ from a list?  
Why are dictionaries particularly useful when working with structured data like sales records?

A dictionary in Python is a data structure that stores data as key–value pairs, where each key is unique and maps to a specific value. It allows data to be accessed using meaningful labels rather than numeric positions.

A dictionary differs from a list in several ways. A list stores elements in an ordered sequence and items are accessed using their index positions. Lists are best suited for collections of similar items. In contrast, a dictionary stores data using keys instead of indices, making it easier to represent relationships between pieces of data. Dictionaries are also more suitable when each item has multiple attributes.

Dictionaries are useful when working with structured data like sales records because each record often contains multiple fields such as order ID, city, category, and amount. Using a dictionary allows each of these fields to be clearly labelled, making the data more readable and easier to work with. It also improves efficiency when retrieving or updating specific pieces of information, since values can be accessed directly using their keys instead of searching through a list.

# Part 3 — Sets

*Sets store only unique values and are extremely fast for membership testing and comparing groups. In data work they are ideal for deduplication and spotting discrepancies between datasets.*

## Question 3.1 — Written

What is a set in Python? List **three** operations that sets support that make them useful in data analysis. Give a brief example use case for each.

A set in Python is an unordered collection of unique elements, meaning it does not allow duplicate values. Sets are useful when working with data where uniqueness and membership are important.

Three key operations supported by sets that are useful in data analysis are:

##union
This operation combines elements from two sets, removing duplicates.
use case: combining customer lists from two different sales regions to get a list of all unique customers.
##intersection
This operation returns only the elements that are common to both sets.
use case: identifying customers who made purchases in both Lagos and Abuja.
##difference
This operation returns elements that are in one set but not in another.
use case: finding customers who purchased in Lagos but not in Port Harcourt.

These operations make sets powerful for tasks such as data cleaning, comparison, and identifying relationships between datasets.

## Question 3.2 — Code

Two DataTrove analysts based in Lagos and Abuja have each recorded the product categories they processed this week:

```python

                  "Food & Drink", "Electronics", "Clothing"]
abuja_analyst  = ["Sports & Outdoors", "Electronics", "Beauty & Health",
                  "Clothing", "Food & Drink", "Toys & Games", "Clothing"]
```

Using sets, find and print:
1. The **unique** categories each analyst processed (no duplicates)
2. Categories processed by **both** analysts (intersection)
3. Categories processed by the **Lagos analyst only** (difference)
4. Categories processed by the **Abuja analyst only** (difference)
5. **All** unique categories processed between both analysts (union)
6. Categories processed by **one analyst but not both** (symmetric difference)

````python
lagos_analyst = ["Electronics", "Clothing", "Electronics", "Home & Garden",
                 "Food & Drink", "Electronics", "Clothing"]

abuja_analyst = ["Sports & Outdoors", "Electronics", "Beauty & Health",
                 "Clothing", "Food & Drink", "Toys & Games", "Clothing"]

# Convert lists to sets (removes duplicates automatically)
lagos_set = set(lagos_analyst)
abuja_set = set(abuja_analyst)

# 1. Unique categories each analyst processed
print("Unique Lagos categories:", lagos_set)
print("Unique Abuja categories:", abuja_set)

# 2. Categories processed by both analysts (intersection)
print("Both analysts:", lagos_set & abuja_set)

# 3. Lagos only (difference)
print("Lagos only:", lagos_set - abuja_set)

# 4. Abuja only (difference)
print("Abuja only:", abuja_set - lagos_set)

# 5. All unique categories (union)
print("All categories:", lagos_set | abuja_set)

# 6. Categories processed by one but not both (symmetric difference)
print("Either but not both:", lagos_set ^ abuja_set)
````

# Part 4 — List Comprehensions

*List comprehensions are a concise, Pythonic way to build lists. They are faster than loops and are used constantly in data transformation. Once you understand them, your code will become significantly cleaner.*

## Question 4.1 — Written

What is a list comprehension? Write out the general syntax and explain each part.  
Rewrite the following loop as a list comprehension and explain what changed:

```python
results = []
for x in range(10):
    if x % 2 == 0:
        results.append(x ** 2)
```

A list comprehension is a concise way of creating a list in Python using a single line of code. It combines a loop, a condition, and an expression into one statement.

General syntax:
[expression for item in iterable if condition]

Explanation:
- expression: the operation performed on each item (e.g. x**2)
- item: the variable used in the loop (e.g. x)
- iterable: the sequence being looped over (e.g. range(10))
- condition: filters items to include only those that meet the requirement (e.g. x % 2 == 0)

The list comprehension version of the loop is:
results = [x**2 for x in range(10) if x % 2 == 0]

What changed:
The loop, condition, and append() method are combined into one line. There is no need to create an empty list first, and the code becomes shorter and easier to read.

````python
results = [x**2 for x in range(10) if x % 2 == 0]
print(results)
````

## Question 5.1 — Written

What is a lambda function? How does it differ from a regular `def` function?  
When would you use one over the other?

A lambda function is a small, anonymous function in Python that is defined using the lambda keyword. It is used to create simple functions in a single line without giving them a name.

General syntax:
lambda arguments: expression

A lambda function differs from a regular def function in several ways. A lambda function does not have a name, while a def function is named. A lambda function can only contain one expression, while a def function can contain multiple lines of code, statements, and more complex logic. Lambda functions are written in a single line, while def functions can span multiple lines.

Lambda functions are best used for short, simple operations that are used temporarily, such as inside functions like map(), filter(), or sorted(). Regular def functions are used when the logic is more complex, needs multiple steps, or when the function will be reused.

The purpose of try / except in Python is to handle errors during program execution so that the program does not crash. It allows you to try running some code and catch any errors that occur, so you can handle them properly, especially when working with messy or unpredictable data.

Three common errors when processing messy datasets are:

## ValueError
This occurs when a value is of the correct type but not in the expected format.
For example, converting "abc" to a float.
Cause: invalid or dirty data in numeric fields.
## KeyError
This happens when trying to access a key that does not exist in a dictionary.
For example, accessing order["price"] when "price" is missing.
Cause: missing or inconsistent data fields.
## TypeError
This occurs when an operation is applied to incompatible data types.
For example, adding a string and an integer ("10" + 5).
Cause: mixing data types without proper conversion.

## Question 6.2 — Code

The following raw price values were pulled from a messy data entry form in Ibadan:

```python
raw_prices = ["49.99", "134.97", "N/A", "", "29.99", "abc", "₦89.99", "12.50", None, "0"]
```

Write a function called `safe_convert_price` that:
- Tries to convert the value to `float`
- If it fails for any reason, returns `None` and prints: `"Warning: could not convert '[value]' to float"`
- If the value converts but is 0 or negative, returns `None` and prints: `"Warning: invalid price '[value]'"`

Then use a list comprehension with your function to build `clean_prices` (excluding `None`) and print the result.

````python
raw_prices = ["49.99", "134.97", "N/A", "", "29.99", "abc", "₦89.99", "12.50", None, "0"]

def safe_convert_price(value):
    try:
        price = float(value)
        if price <= 0:
            print(f"Warning: invalid price '{value}'")
            return None
        return price
    except:
        print(f"Warning: could not convert '{value}' to float")
        return None

# Build clean_prices list:
clean_prices = [p for p in (safe_convert_price(v) for v in raw_prices) if p is not None]

print(clean_prices)
````

## Question 6.3 — Code

Write a function called `safe_divide` that:
- Takes two numbers `a` and `b`
- Returns `a / b`

- Handles `TypeError` (e.g. if `b` is a string) and returns `None` with a different warning

Test it with at least four different inputs covering the normal case and both error cases.

````python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Warning: cannot divide by zero")
        return None
    except TypeError:
        print("Warning: invalid input types for division")
        return None

# Test cases:
print(safe_divide(10, 2))      # normal case
print(safe_divide(10, 0))      # ZeroDivisionError
print(safe_divide(10, "2"))    # TypeError
print(safe_divide("10", 2))    # TypeError
````

## Question 6.4 — Code ⭐ Challenge

Write a robust function called `process_order_row` that takes a dictionary representing a raw, potentially dirty data row and:

1. Safely extracts and converts each field using `try/except`
2. Converts `quantity` to `int`, `unit_price` and `discount_pct` to `float`
3. Calculates `total_amount` if the conversions succeeded
4. Returns a cleaned dictionary with an `"errors"` key listing any fields that failed
5. Returns `"errors": []` if the row was clean

Test it on one clean row and two rows with different problems, and print the result of each.

````python
def process_order_row(row):
    cleaned = {}
    errors = []

    # Copy simple fields
    cleaned["order_id"] = row.get("order_id")
    cleaned["city"] = row.get("city")

    # Quantity
    try:
        qty = int(row.get("quantity"))
        cleaned["quantity"] = qty
    except:
        cleaned["quantity"] = None
        errors.append("quantity")

    # Unit price
    try:
        price = float(row.get("unit_price"))
        cleaned["unit_price"] = price
    except:
        cleaned["unit_price"] = None
        errors.append("unit_price")

    # Discount %
    try:
        discount = float(row.get("discount_pct"))
        if discount < 0 or discount > 100:
            raise ValueError
        cleaned["discount_pct"] = discount
    except:
        cleaned["discount_pct"] = None
        errors.append("discount_pct")

    # Calculate total_amount (only if no errors in required fields)
    if all([
        cleaned.get("quantity") is not None,
        cleaned.get("unit_price") is not None,
        cleaned.get("discount_pct") is not None
    ]):
        total = cleaned["quantity"] * cleaned["unit_price"]
        total_after_discount = total * (1 - cleaned["discount_pct"] / 100)
        cleaned["total_amount"] = total_after_discount
    else:
        cleaned["total_amount"] = None

    cleaned["errors"] = errors
    return cleaned


# Test rows:
clean_row = {"order_id": "DT-001", "city": "Lagos",  "quantity": "3",   "unit_price": "49.99",  "discount_pct": "10"}
bad_row_1 = {"order_id": "DT-002", "city": "Abuja",  "quantity": "abc", "unit_price": "29.99",  "discount_pct": "5"}
bad_row_2 = {"order_id": "DT-003", "city": "Kano",   "quantity": "2",   "unit_price": None,     "discount_pct": "200"}

print(process_order_row(clean_row))
print(process_order_row(bad_row_1))
print(process_order_row(bad_row_2))
````
