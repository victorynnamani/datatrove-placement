## Question 1.5

What is the difference between `print(name)` and `name` when typed in a Jupyter Notebook cell? When would each be more useful?

### Answer:

In a Jupyter Notebook, typing `name` on its own will display the value of the variable only if it is the last line in the cell. This is because Jupyter automatically outputs the result of the final expression.

On the other hand, `print(name)` explicitly displays the value of the variable regardless of where it appears in the code. It can be used multiple times within the same cell.

`print(name)` is more useful when you want to display multiple outputs or format your results clearly. Simply typing `name` is more useful for quick inspection when working interactively.

## Question 4.1

What is the difference between `if`, `elif`, and `else`? Write a brief explanation and give a real-world data scenario where you would need all three.

### Answer:

In Python, `if`, `elif`, and `else` are used to make decisions in code.

- `if` is used to test a condition. If the condition is true, the code inside it runs.
- `elif` means "else if" and is used to check another condition if the previous one was false.
- `else` runs if none of the previous conditions are true.

A real-world data example is classifying product prices. For instance, if a product is below £10, it is labelled as "Budget". If it is between £10 and £50, it is "Standard". If it is between £50 and £100, it is "Premium". Otherwise, it is "Luxury".

````python
# ## Question 4.2
# This code classifies a unit price into a pricing tier
# (Budget, Standard, Premium, or Luxury) using if/elif/else conditions
# and prints the corresponding price tier

unit_price = 75  

if unit_price < 10:
    tier = "Budget"
elif unit_price < 50:
    tier = "Standard"
elif unit_price < 100:
    tier = "Premium"
else:
    tier = "Luxury"

print(f"Price tier: {tier}")
````

## Question 5.1

What is the difference between a `for` loop and a `while` loop? Give one example use case for each in a data analysis context.

### Answer:

A `for` loop is used to iterate over a sequence (such as a list) and runs a fixed number of times based on the items in that sequence.

A `while` loop runs as long as a condition remains true, and it is used when the number of iterations is not known in advance.

In data analysis, a `for` loop can be used to go through a list of sales values and process each one. A `while` loop can be used to keep processing data until a certain condition is met, such as continuing to read transactions until no more records are left.

````python
# ## Question 5.2
# This code loops through a list of sales figures,
# prints each value, and labels it as "Above target" or "Below target"
# based on whether it meets the £1000 threshold

# Step 1: Define the list
sales = [1200, 450, 3800, 290, 5100, 620, 4400, 180, 2750, 990]

# Step 2: Loop through the list
for sale in sales:
    print(sale)
    
    if sale >= 1000:
        print("Above target")
    else:
        print("Below target")
````

````python
# ## Question 5.3
# This code uses a loop to manually calculate the total, average,
# highest, and lowest sales values without using built-in functions
# and prints all the computed results

# Step 1: Define the list
sales = [1200, 450, 3800, 290, 5100, 620, 4400, 180, 2750, 990]

# Step 2: Initialise variables
total = 0
highest = sales[0]
lowest = sales[0]

# Step 3: Loop
for sale in sales:
    total += sale
    
    if sale > highest:
        highest = sale
        
    if sale < lowest:
        lowest = sale

# Step 4: Calculate average
average = total / len(sales)

# Step 5: Print results
print("Total:", total)
print("Average:", round(average, 2))
print("Highest:", highest)
print("Lowest:", lowest)
````

````python
# ## Question 5.5
# This code simulates account transactions using a while loop,
# deducts a random amount from the balance each time,
# prints the remaining balance, and stops when the balance goes below zero
# while counting and reporting the number of transactions

# Step 1: Import module
import random

# Step 2: Set seed
random.seed(42)

# Step 3: Initialise variables
balance = 1000
count = 0

# Step 4: Loop until balance is gone
while balance >= 0:
    deduction = random.randint(50, 200)
    balance -= deduction
    count += 1
    
    print("Balance:", balance)

# Step 5: Final output
print("Account exhausted after", count, "transactions")
````

## Question 6.1

What is a function in Python? What are the benefits of using functions instead of writing the same code multiple times? Give a data-related example.

### Answer:

A function in Python is a reusable block of code that performs a specific task. It helps organise programs into smaller and more manageable parts.

One major benefit of using functions is that they reduce repetition. Instead of writing the same code multiple times, a function allows you to write it once and reuse it whenever needed. This makes code cleaner, easier to understand, and easier to maintain. Functions also help in debugging because errors can be traced to a specific function.

In data analysis, for example, a function can be created to clean text data by removing extra spaces and converting all text to lowercase. This function can then be applied to many rows of data instead of rewriting the cleaning steps repeatedly.

````python
#Question 6.2
#  This function calculates the final total cost after applying a percentage discount to a purchase

def calculate_total(unit_price, quantity, discount_pct):
    total = unit_price * quantity
    discount = total * (discount_pct / 100)
    final_total = total - discount
    return round(final_total, 2)

# Test your function
print(calculate_total(50, 2, 10))
print(calculate_total(100, 5, 20))
print(calculate_total(49.99, 3, 10))
````

````python
#Question 6.3
#  This function classifies product prices into categories such as Budget, Standard, Premium, or Luxury

def classify_price(unit_price):
    if unit_price < 10:
        return "Budget"
    elif unit_price < 50:
        return "Standard"
    elif unit_price < 100:
        return "Premium"
    else:
        return "Luxury"

prices = [4.99, 12.99, 34.99, 49.99, 74.99, 129.99, 9.99, 54.99]

for price in prices:
    print(price, "-", classify_price(price))
````

````python
#Question 6.4
# This function counts the number of vowels (a, e, i, o, u) in a given text

def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

result = count_vowels("DataTrove Analytics")
print("There are", result, "vowels in 'DataTrove Analytics'")
````

````python
#Question 6.5
#  This function checks whether an order contains valid data and returns a list of any issues found

def validate_order(order):
    issues = []
    valid_regions = ["North", "South", "East", "West"]

    if order["quantity"] <= 0:
        issues.append("Invalid quantity")

    if order["unit_price"] <= 0:
        issues.append("Invalid unit price")

    if order["discount_pct"] < 0 or order["discount_pct"] > 100:
        issues.append("Invalid discount percentage")

    if order["region"] not in valid_regions:
        issues.append("Invalid region")

    if order["return_flag"] not in ["Y", "N"]:
        issues.append("Invalid return flag")

    return issues


# Test with correct order
order1 = {
    "order_id": "DT-10001",
    "product_name": "Laptop",
    "quantity": 2,
    "unit_price": 500,
    "discount_pct": 10,
    "region": "North",
    "return_flag": "N"
}

print(validate_order(order1))


# Test with broken order
broken_order = {
    "order_id": "DT-00000",
    "product_name": "Phone",
    "quantity": 0,
    "unit_price": -10,
    "discount_pct": 150,
    "region": "Unknown",
    "return_flag": "X"
}

print(validate_order(broken_order))
````

## Question 7.1

What is the difference between `append()` and `extend()` in Python lists? Write your explanation and show a code example of each.

### Answer:

The `append()` method adds a single element to the end of a list. If you append another list, it will be added as one item (a nested list).

The `extend()` method adds each element of another list to the original list individually. It merges the two lists into one.

Example:

```python
# append example
list1 = [1, 2, 3]
list1.append([4, 5])
print(list1)  # Output: [1, 2, 3, [4, 5]]

# extend example
list2 = [1, 2, 3]
list2.extend([4, 5])
print(list2)  # Output: [1, 2, 3, 4, 5]

````python
#Question 7.2
#  This code cleans product category names to title case, removes duplicates while preserving original order,
# sorts the final list alphabetically, and prints both the cleaned list and the number of unique categories

raw_categories = ["electronics", "CLOTHING", "home & garden", "Electronics",
                  "sports & outdoors", "CLOTHING", "food & drink", "electronics",
                  "Beauty & Health", "toys & games", "books & stationery", "CLOTHING"]

cleaned_categories = []

# Step 1 & 2: clean to title case and remove duplicates (preserve order)
for category in raw_categories:
    cleaned = category.strip().title()
    if cleaned not in cleaned_categories:
        cleaned_categories.append(cleaned)

# Step 3: sort alphabetically
cleaned_categories.sort()

# Step 4: print cleaned list
print("Cleaned categories:", cleaned_categories)

# Step 5: print number of unique categories
print("Number of unique categories:", len(cleaned_categories))
````

````python
# ## Question 7.4
# This code analyses a list of sales transactions by:
# 1. printing all Electronics transactions
# 2. calculating total revenue
# 3. finding the highest value transaction
# 4. counting how many transactions are over £50

transactions = [
    {"product": "Wireless Headphones", "category": "Electronics", "total": 134.97},
    {"product": "Yoga Mat",            "category": "Sports & Outdoors", "total": 27.99},
    {"product": "Smart Watch",         "category": "Electronics", "total": 259.98},
    {"product": "Throw Blanket",       "category": "Home & Garden", "total": 29.99},
    {"product": "Protein Powder",      "category": "Food & Drink", "total": 59.98},
    {"product": "Bluetooth Speaker",   "category": "Electronics", "total": 34.99},
    {"product": "Hiking Backpack",     "category": "Sports & Outdoors", "total": 129.98},
    {"product": "Face Moisturiser",    "category": "Beauty & Health", "total": 18.99},
]

# 1. Print Electronics transactions
print("Electronics transactions:")
for t in transactions:
    if t["category"] == "Electronics":
        print(t)

# 2. Calculate total revenue
total_revenue = 0
for t in transactions:
    total_revenue += t["total"]

print("Total revenue:", round(total_revenue, 2))

# 3. Find highest transaction
highest = transactions[0]
for t in transactions:
    if t["total"] > highest["total"]:
        highest = t

print("Highest transaction:", highest["product"], "-", highest["total"])

# 4. Count transactions above £50
count = 0
for t in transactions:
    if t["total"] > 50:
        count += 1

print("Transactions above £50:", count)
````
